1. The Two Pillars of Pandas
In Pandas, almost everything is either a Series or a DataFrame:

DataFrame: The whole table (like a 2D array or an Excel sheet). It has rows and columns.

Series: A single column (like a 1D array). A DataFrame is essentially a collection of Series objects that share the same index.
