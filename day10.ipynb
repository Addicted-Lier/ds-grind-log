Day 10: Correlation ≠ Causation
This is the most famous trap in all of data science.

Correlation means two things move together. (When A goes up, B goes up).
Causation means A actually causes B to happen.

The Classic "Ice Cream" Trap
Statistics show that as Ice Cream Sales increase, the number of Shark Attacks also increases. They are perfectly correlated!

The Question: Does eating ice cream make sharks hungry for humans?
The Reality: No. There is a "Hidden Variable" (Confounder)—Summer/Heat. When it's hot, people buy ice cream AND they go swimming in the ocean.

Learning Activity: The "Broken Conclusion" Challenge
Suppose you find a dataset showing that people who own more than 5 books have higher life expectancy.

Does buying more books cause your heart to beat longer?

What could be the "Hidden Variable" that explains why book owners live longer? (Think about wealth, education, or lifestyle).
