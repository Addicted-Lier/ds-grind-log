Day 6: Advanced Sorting and the "Top-K" Challenge
Now that you have your grouped data, it’s usually a mess. To make it useful, we need to sort it. This is very similar to "sorting" in Competitive Programming (CP), where you often have to sort by one criteria, then break ties with another.

1. Basic Sorting
Use .sort_values().

Ascending (Smallest to Largest): df.sort_values('rating')

Descending (Largest to Smallest): df.sort_values('rating', ascending=False)



```python
df=pd.read_csv("games3.csv")
df.sort_values('rating')
```

```python
df.sort_values('rating', ascending=False)
```

2. Multi-Level Sorting (Breaking Ties)
What if two games have the same rating? You might want to sort by Rating first (high to low), and then by Year (newest to oldest). df.sort_values(['rating', 'release_year'], ascending=[False, False])

```python
df.sort_values(['rating', 'release_year'], ascending=[True,False])

# means:

# rating → ascending (small → big)

# if rating ties → release_year descending (new → old)

# Priority 1 → rating
# Priority 2 → release_year
```

3. The "Stuck" Point: Top-K per Group
This is where most people get stuck. If I ask for the top 2 best-selling games in EVERY genre, a simple sort won't work because it sorts the whole table.

The Trick: We use .nlargest(). df.groupby('genre')['units_sold_m'].nlargest(2)

```python
df.groupby('genre')['units_sold_m'].nlargest(2)
```
