Day 9: Distributions and Skewness
we need to understand the shape of the data. This is called the Distribution.

```python
1. The Normal Distribution (The Bell Curve)
In a perfect world, your data is symmetrical. The Mean, Median, and Mode are all the same.
Example: Human heights. Most people are average, with a few very tall or very short people.
```

. Skewed Distributions
When your data has a "tail" pulling it in one direction:

Right Skewed (Positive): The tail is on the right. The Mean is pulled higher than the Median (like our Salary example!).

Left Skewed (Negative): The tail is on the left. The Mean is pulled lower than the Median.

In your Housing Example (Right Skew):
The Crowd: Most houses are clustered at $500k.

The Outliers: Those $10M mansions are way out to the Right on the number line.

The Tail: Because those mansions are so far away, they create a long "tail" stretching to the Right.

The Pull: The Mean is sensitive. It sees those $10M numbers and gets "pulled" toward them.

The Result: The Mean moves to, say, $2M, while the Median stays at $500k.

Conclusion: Mean > Median = Right Skewed.
