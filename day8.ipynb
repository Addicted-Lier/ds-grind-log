Day 8: Thinking Before Plotting
Today is about "Number-only EDA." Before we make pretty charts, we need to describe the data using statistics.

Key Tools for Day 8:

df.describe(): The "Magic Wand." It gives you the mean, median (50%), std deviation, min, and max for every number column.

df['column'].value_counts(): Tells you the frequency of items (e.g., "How many Finance apps vs Game apps?").

df['column'].unique(): Lists all the distinct categories.

```python
The Power of df.describe()
This one command is your best friend. It gives you the "Five-Number Summary."

Mean: The average. (Can be tricked by one really high number!)

50% (Median): The middle value. (The "Truth-teller" when there are outliers).

Std (Standard Deviation): How spread out the numbers are. If this is high, your data is "wild."
```

Learning Activity: The "CEO" Paradox
